```python
names = []
ids = []
duration = []
album_type = []
for track in tracks['items']:
    names.append(track['track']['name'])
    ids.append(track['track']['id'])
    duration.append(track['track']['duration_ms']/1000)
    album_type.append(track['track']['album']['album_type'])
```

```python
import pandas as pd
df = pd.DataFrame({'name':names , 'id':ids , 'duration': duration , 'album_type' : album_type})
```
